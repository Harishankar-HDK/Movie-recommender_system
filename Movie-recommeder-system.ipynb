```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
#Now to avoid the hectic situation we merge the two datsets like as below
movies = movies.merge(credits,on='title')
```

```python
movies.info()
```

```python
#We dont need all these columns so we just take the required columns like
# genres
# id
# keywords
# title
# cast
# crew
# overview
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
#IN the dataset genre column keywords,cast and crew column is in a weird format we need to convert it into a suitable format
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# here we need to convert this into [Action,Adventure,Fantasy,Science-Fiction]
# to do that we will create a function as below
```

```python
import ast
#This library import ast which have function called literal_eval which will help to convert string into list
# ex: '[1,2,3]' this is string list with the help of literal_eval we can convert it into [1,2,3]
def converter(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
converter('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(converter)
```

```python
movies.head()
```

```python
movies['keywords'][0]
#Here keyword is also same like the genres so we need only the names so we can use the same function for keyword column also
```

```python
movies['keywords'] = movies['keywords'].apply(converter)
```

```python
movies.head()
```

```python
#Next the cast column ,in cast column there is name of all actors who have acted ,we can take all the names so
#We prefer taking the first 3 names ,the function is similar to the function we have used for genre and keyword but
# A slight difference we does not need to iterate through all the names we need only first 3 names so we can create function as follow
def convert_3(obj):
    L=[]
    counter = 0 #This is use to know whether we have reached the 3 names,if counter=3 we need to break from the loop
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter = counter+1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#Next a bit challenging is Crew column, we need to fetch only director name we can ignore the rest
#The function is similar to the one that we have used before but only difference is we have to fetch only the director name
#so the function is as follow
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#Now we need to convert the overview into a list for convenience
movies['overview'] = movies['overview'].apply(lambda x:x.split())
#This will convert the overview into a list
movies.head()
```

```python
#Now an important thing to do is, there is many name in the dataset like Sam worthington,Sam mendes etc
#Sam worthington is name of one person and Sam mendes is name of other person
#if the model return Sam ,Then it will confused whether it is Sam Mendes or Sam Worthington so to avoid this 
#We can remove the space between the names so to do that we do the following
movies['genres'] = movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Now overview,genres,keywords,cast and crew is in list we can concatenate all these into one column named tag
# And assign it to a new variable
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
movies.head()
```

```python
# Now we can create a new dataframe which contain only movie_id,title and tags
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
#Now we can convert the tags list into string with a simple function as follows
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
#Now for convenience we can convert all the tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
    y=[]
    for i in text.split():
        y.append(lemmatizer.lemmatize(i))
   
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(lemma)
#We have used stemming to avoid repetiton of the words
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features = 5000,stop_words='english')
#We are using countvectorizer to convert the string 'tags' into vectors for similarity check
# we cannot check for similarity between movies with the help of the strings so we are using vectors to check the similarity
# Thats why we have used countvectorizer
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
# Now our main goal is movie recommendation so we have to check similarity between the movies
# we need to compare each movie with the rest of the movie to check the similarity
# we have already converted the tags of the movies to vector for similarity checking convenince
# For similarity check we will use cosine similarity function instead of eucledian distance
# Because in higher dimension eucledian distance is not realiable
# as distance increase similarity decrase 
# In cosine similarity each movie is checked with rest of the 5000 movies for similarity and it will result an array
# I will show the working in below
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
#It gives 4806,4806 because we need to check each movie with the rest of the movie
#That is 1st movie with 4086 movies second movie with 4086 movies and so on therefore there will be total 4086x4086 checking
```

```python
#Now we need to define a function named recommed to for recommending movies based on similarity
#That is,when we give a movie the function should return top 5 similar movies corresponding to the given movie
#For example if we give avatar as the movie then the function should return the top 5 movies similar to avatar 
#for that we need movie index,then give the index of the movie to cosine similarity to check with the rest of the movies
#Then the step will be explained after creating the function
```

```python
new_df[new_df['title']=='Avatar']
```

```python
new_df[new_df['title']=='Avatar'].index[0]
```

```python
list(enumerate(similarity[0]))
#Here index and the similarity value can be seen with the help of the enumerate
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
#Here key=lambda x:x[1] is a function which tells the sorted function to sort the values based on the similarity value not index
#That is we can see first value represent the index and second value represent the similarity value
#ex: (0,1.00)
    #(1,0.083) etc
# here 0 and 1 represent the index value and 1.00 and 0.083 represent the similarity value
# by default sorted function sort based on the index value but when we use to lambda function
# lambda x:x[1] we tell the sorted function to sort base on the similarity value
# x:x[1] tell to sort by similarity value x:x[0] will tell to sort by index value
# [1:6] represent we want first 5 movie values not all 
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0] #This will return the index of any movie given
    distances = similarity[movie_index] #This will do the cosine similarity so that distance between movie can be get
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #first 5 movies will be returned
    for i in movie_list:
        print(new_df.iloc[i[0]].title) #This will print the name of the movie that is similar to the given movie
        print(i[1])
        print(i[0])
```

```python
recommend('Pirates of the Caribbean: The Curse of the Black Pearl')
```

```python
recommend('Batman Begins')
```
